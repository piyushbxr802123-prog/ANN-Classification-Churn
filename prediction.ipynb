```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model=load_model('model.h5')

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
    'Geography': 'France'
}
```

```python
geo_encoded=label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=label_encoder_geo.get_feature_names_out(['Geography']))
```

```python
input_df=pd.DataFrame([input_data])
```

```python
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
```

```python

input_df=pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
input_df
```

```python
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
import numpy as np

# Load saved weights
data = np.load("model_weights.npz")
W = [data[k] for k in data.files]

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def predict(X):
    a = relu(X @ W[0] + W[1])       # hidden layer 1
    a = relu(a @ W[2] + W[3])       # hidden layer 2
    a = sigmoid(a @ W[4] + W[5])    # output layer
    return a
```

```python


prediction = predict(input_scaled)          # ✅ use our numpy function
prediction_proba = float(prediction[0][0])  # ✅ convert to normal number
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is unlikely to churn.")
```
